```python
def get_val(row_data):
    if row_data.find('li'):
            return [li.get_text(' ', strip=True).replace('\xa0',' ') for li in row_data.find_all('li')]
    else:
        return row_data.find('td').get_text(' ', strip=True).replace('\xa0',' ')
    
    
movie_info = {}

for index , rows in enumerate(tb_rows):
    if index == 0:
        movie_info['title'] = rows.find('th').get_text(' ', strip=True)
    elif index == 1:
        continue
    else:
        key = rows.find('th').get_text(' ', strip=True).replace('\xa0',' ')
        val = get_val(rows)
        movie_info[key] = val

# for removing [1] that refers to reference links
for i,n in movie_info.items():
    if str(n).find('[1]'):
        str(n).strip('[1]')
    else:
        continue

movie_info
```
